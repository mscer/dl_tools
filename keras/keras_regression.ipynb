```python
def plot_curve(history_val):
    epochs = range(1, len(history_val['acc']) + 1)
    plt.plot(epochs, history_val['loss'], 'bo', label='Training loss')
    plt.plot(epochs, history_val['val_loss'], 'b', label='Validation loss')
    plt.title('Training and validation loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    epochs = range(1, len(history_val['acc']) + 1)
    plt.plot(epochs, history_val['acc'], 'ro', label='Training acc')
    plt.plot(epochs, history_val['val_acc'], 'r', label='Validation acc')
    plt.title('Training and validation acc')
    plt.xlabel('Epochs')
    plt.ylabel('acc')
    plt.legend()
```

# 1 data&feat

```python
from keras.datasets import boston_housing
```

```python
(train_data, train_targets), (test_data, test_targets) = boston_housing.load_data()
```

```python
mean = np.mean?
```

```python
mean = np.mean(train_data,axis=0)
```

```python
std = np.std(train_data,axis=0)
```

```python
x_train = (train_data-mean)/std
x_test = (test_data-mean)/std
```

# 2 model

```python
nn = keras.models.Sequential()
nn.add(keras.layers.Dense(units=64,activation='relu',input_shape=(13,)))
nn.add(keras.layers.Dense(units=64,activation='relu'))
nn.add(keras.layers.Dense(units=1))
```

```python
nn.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

```python
history = nn.fit(x_train,train_targets,batch_size=128,epochs=500,validation_split=0.2)
```

```python
def plot_curve(history_val,flag='1'):
    epochs = range(1, len(history_val['mean_squared_error']) + 1)
    if '1' == flag:
        plt.plot(epochs, history_val['loss'], 'bo', label='Training loss')
        plt.plot(epochs, history_val['val_loss'], 'b', label='Validation loss')
        plt.title('Training and validation loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()
    else:
        plt.plot(epochs, history_val['mean_squared_error'], 'ro', label='Training mse')
        plt.plot(epochs, history_val['val_mean_squared_error'], 'r', label='Validation mse')
        plt.title('Training and validation mse')
        plt.xlabel('Epochs')
        plt.ylabel('mse')
        plt.legend()
```

```python
plot_curve(history.history)
```

```python
plot_curve(history.history,1)
```

# 3 eval

```python
nn.evaluate(x_test,test_targets)
```
